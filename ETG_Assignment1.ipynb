```python
fd=open("Record.json",'r')
txt=fd.read()
fd.close()
Record=json.loads(txt)
x=0

while (x==0):
  print("---------------------------------------------------------")
  print("*******************************************************")
  print("Choose the function you want to perform")
  choice=input("Press 1 to browse inventory by category\nPress 2 to purchase an item from inventory \nPress 3 to Add an item to the inventory \nPress 4 to delete an item from the inventory \nPress 5 to update inventory stock \nPress 'Any Other' key to exit: \n")
  print("*******************************************************")
  print("---------------------------------------------------------\n")
  
  if (choice=='1'):
    cat_lst=[]
    for i in Record:
      cat=Record[i]['Category']
      cat_lst.append(cat)
    unique_category=set(cat_lst)
    n=1
    for i in unique_category:
      print(i)
      n=n+1
    z=0
    choice=input("Pick category: ")
    for i in Record:
      if (Record[i]['Category']==choice):
        print(i,":",Record[i])
     
  elif (choice=='2'):       
    p_id=(input("Enter the product id of the item to be purchased: "))
    p_units=int(input("Enter the number of items you want to purchase: "))
    if (p_units > Record[p_id]['AvailableItems']):
      print("*******************************************************")
      print("Sorry we do not have the required no. of units in stock")
      print("*******************************************************")
    else:
      print("*******************************************************")
      print("Product Name: ",Record[p_id]['ProductName'])
      print("Quantity: ",p_units)
      print("Total Price: ",Record[p_id]['PriceOfSingleUnit']*p_units)
      print("*******************************************************")
      Record[p_id]['AvailableItems']=Record[p_id]['AvailableItems']-p_units
      js=json.dumps(Record)
      fd=open("Record.json",'w')
      fd.write(js)
      fd.close()
      sales[len(sales)+1]={'ProductId':p_id,'ProductName':Record[p_id]['ProductName'],'Quantity':Record[p_id]['Quantity'],'TotUnits':p_units,'Price':Record[p_id]['PriceOfSingleUnit']*p_units}
      sale=json.dumps(sales)
      fd2=open("Sales.json",'a')
      fd2.write(sale)
      fd2.close()

  
  elif (choice=='3'):
    print("Enter the following details of the new item")
    np_id=(input("Product ID: "))
    np_name=input("Product Name: ")
    np_Category=input("Product Category: ")
    np_Quantity=input("Product Quantity: ")
    np_AvailableItems=int(input("Total no. of Available items: "))
    np_Price=float(input("Enter the price of single unit of the product"))
    Record[np_id]={'ProductName':np_name,'Category':np_Category,'Quantity':np_Quantity,'AvailableItems':np_AvailableItems,'PriceOfSingleUnit':np_Price}
    js=json.dumps(Record)
    fd=open("Record.json",'w')
    fd.write(js)
    fd.close()
    print("*******************************************************")
    print("The Item",Record[np_id]['ProductName']," has been successfully added to the inventory")
    print("*******************************************************")
  
  elif (choice=='4'):
    p_id=(input("Enter the ID of the product you want to delete"))
    p_name=Record[p_id]['ProductName']
    del Record[p_id]
    js=json.dumps(Record)
    fd=open("Record.json",'w')
    fd.write(js)
    fd.close()
    print("*******************************************************")
    print("The Item ",p_name," has been successfully Removed from the inventory")
    print("*******************************************************")
  
  elif (choice=='5'):
    up_id=(input("Enter the ID of the product you want to update: "))
    up_stock=int(input("Enter the number of units you want to add to the stock: "))
    Record[up_id]['AvailableItems']= Record[up_id]['AvailableItems']+up_stock
    js=json.dumps(Record)
    fd=open("Record.json",'w')
    fd.write(js)
    fd.close()
    print("*******************************************************")
    print("The stock has been updated successfully for the item ",Record[up_id]['ProductName'])
    print("*******************************************************")
  
  else:
    x=1
```
